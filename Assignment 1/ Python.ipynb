# NO.3 PROGRAM HITUNG BMI

3. Buatlah program untuk menghitung BMI!

```python
# Fungsi untuk menghitung BMI
def hitung_bmi(berat_badan, tinggi_badan_cm):
    tinggi_badan_m = tinggi_badan_cm / 100
    bmi = berat_badan / (tinggi_badan_m ** 2)
    return bmi

# Input berat badan dan tinggi badan
berat_badan = float(input("Masukkan berat badan (kg): "))
tinggi_badan_cm = float(input("Masukkan tinggi badan (cm): "))

# Menghitung BMI
bmi = hitung_bmi(berat_badan, tinggi_badan_cm)

# Menampilkan hasil BMI
print(f"Nilai BMI Anda adalah: {bmi:.2f}")
```

# NO.5 STATUS BMI

5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.

*   Normal: 18.5-24.9
*   Overweight: 25.0-29.9
*   Obesity: 30.0-34.9
*   Underweight: < 18.5




```python
def kategori_bmi(BMI):
  if BMI < 18.5:
    return 'Underweight'
  elif 18.5 <= BMI <= 24.9:
    return 'Normal'
  elif 25.0 <= BMI <= 29.9:
    return 'Overweight'
  elif 30.0 <= BMI <= 34.9:
    return 'Obesity'
  else:
    return 'Unknown'

df_g['Status BMI'] = df_g['BMI'].apply(kategori_bmi)
df_g['Status BMI']
```
